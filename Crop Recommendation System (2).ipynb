```python
#Feature Selection
x=df.drop("label",axis=1)  #x=feature,means input
y=df["label"]  #y=target,means output
```

```python
#Display Feature
x.head()
```

```python
#display Target
y.head()
```

```python
#Import train-test split
from sklearn.model_selection import train_test_split
```

```python
#split dataset into training and testing set
#we split the data into training and testing->80%(1760) training and 20%(440) testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42) # test size = 0.2 means we test on 20% of data set and train 80% of dataset and random state=42 means ,splitting will remain same after every run of the notebook
print("x_train:",x_train.shape) # train 1760 rows of the input
print("x_test:",x_test.shape) # test 440 rows of the input
print("y_train:",y_train.shape) # train 1760 rows of the output
print("y_test:",y_test.shape) # test 440 rows of the output
# we split the data for training and testing to evaluate how models performs on unseen data
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler  #standard scale is used for transform the data so that mean=0,SD=1,after scaling all features have comparable ranges
scaler=StandardScaler()   #create a scalar object
x_train=scaler.fit_transform(x_train)  #.fit->learns the mean and SD ,transform->use thode values to scale the training data
x_test=scaler.transform(x_test)  #only use transform for scaling the test data because we dont need new mean and SD
```

```python
#check scaled data
print(x_train.shape)  #shape remains same
print(x_test.shape)
```

```python
#Model training using Random Forest
#Random Forest is a supervides machine learning algorithm that combines the many decision tree and predict the major voting

from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(random_state=42) #it creates the model and ensure the consistent result after every run of the notebook
```

```python
#Train the model
model.fit(x_train,y_train) #fit() train the model and pattern learning
```

```python
#Predict on Test data 
y_pred=model.predict(x_test) # Here is the test data (X_test).Predict which crop each row belongs to and The predictions are stored in the variable y_pred
print(y_pred[:10]) #this shows the first 10 predicted crops
```

```python
#comparison between actual and predictwd crops
comparison=pd.DataFrame({"Actual":y_test.values,"Predicted":y_pred})
comparison.head(10)
```

```python
#Model Evaluation
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)  #accuracy tells us that what percentage of predictions were correct
```

```python
cm=confusion_matrix(y_test,y_pred)  #confusion matrrix compares the actual label and the predicteed label
plt.figure(figsize=(10,8))
sns.heatmap(cm,annot=True,cmap="Blues",fmt="d")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```

```python
print(classification_report(y_test, y_pred))  #classification report tell how well the model predicts the each crop


 #Precision -> Out of all samples predicted as Apple, how many were actually Apple?
 #Recall -> Out of all actual Apple crops, how many did the model correctly identify?
 #F1-score combines Precision and Recall into a single value.If both are high,F1-score is also high.If one is low,F1-score decreases.
 #Support simply means How many test samples belong to that crop?
```

```python
#Save the Trained Model
import joblib
joblib.dump(model,"model.pkl")
joblib.dump(scaler,"scaler.pkl")
```

```python
#Load saved Model
loaded_model = joblib.load("model.pkl")
loaded_scaler = joblib.load("scaler.pkl")
```

```python
#Test the loaded Model
sample = pd.DataFrame([[90, 42, 43, 20.8, 82, 6.5, 202]],columns=["N", "P", "K", "temperature", "humidity", "ph", "rainfall"])
sample_scaled = loaded_scaler.transform(sample)
prediction = loaded_model.predict(sample_scaled)
print("Recommended Crop:", prediction[0])
```
